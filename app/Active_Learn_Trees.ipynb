### Berlin Trees' Dataset Cleaning & Visualizations Based on Species

```python
# Claculating the average height per species
df_average_height = df_cleaning.groupby(['Gattung botanisch'], as_index=False)['Baumhöhe in m'].mean()
df_average_height
```

```python
#Histogram showing the average height by species 
df_average_height.hist()
```

```python
#Pie chart showing the percentage of average height by species 
height_fig = px.pie(df_average_height, values='Baumhöhe in m', names='Gattung botanisch', 
             color_discrete_sequence=px.colors.sequential.RdBu)
height_fig.show()
```
